```python
from aocd import get_data
raw_data = get_data(day=3, year=2025)
voltage_list = raw_data.splitlines()
```

## Part 1

```python
def get_total_voltage(voltage_list, batteries = 2):
    num_list = []
    for line in voltage_list:
        removals = len(line) - batteries
        stack = []
        for char in line:
            if not stack:
                stack.append(char)
                continue
            while stack and removals > 0 and stack[-1] < char:
                stack.pop()
                removals -= 1
            stack.append(char)
        final_stack = stack[:len(stack)-removals]
        num_list.append(int(''.join(final_stack)))
    return sum(num_list)

print(get_total_voltage(voltage_list, batteries=2))
```
