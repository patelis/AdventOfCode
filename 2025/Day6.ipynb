```python
from aocd import get_data
raw_data = get_data(day=6, year=2025)
import re
import numpy as np
```

## Part 1

```python
data = [re.split(" +", s.strip()) for s in raw_data.splitlines() if re.split(" +", s)]
operators = np.array(data[-1])
numbers = np.array(data[:-1], dtype=int).transpose()
part1 = int(np.sum(np.prod(numbers[operators == '*'], axis=1)) + np.sum(numbers[operators == '+']))
print(part1)
```

## Part 2

```python
cols = list(zip(*raw_data.splitlines())) # transpose to columns | transpose trick When you have a list of lists (or strings), zip(*lines) gives you columns instead of rows.

problems = []
current = []
for col in cols:
    if all([i == " " for i in col]):  # separator column
        if current:
            problems.append(current)
            current = []
    else:
        current.append(col)
if current:
    problems.append(current)
    
part2 = 0
for problem in problems:
    operator = problem[0][-1]
    digits_only = [col[:-1] for col in problem[::-1]] # remove operator row and reverse
    numbers = [int("".join(col).replace(' ', '')) for col in digits_only]
    if operator == '+':
        partial_total = sum(numbers)
    else: # operator == '*'
        partial_total = 1
        for n in numbers:
            partial_total *= n
    part2 += partial_total
print(part2)
```
