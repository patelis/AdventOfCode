```python
from aocd import get_data
raw_data = get_data(day=5, year=2025)
raw_fresh, raw_available = raw_data.split('\n\n')
available = [int(i) for i in raw_available.splitlines()]
fresh = [[int(x) for x in li.split("-")] for li in raw_fresh.splitlines()]
```

## Part 1

```python
sum(any(start <= a <= end for start, end in fresh) for a in available)
```

## Part 2

```python
# Sorting and merging intervals
fresh = sorted(fresh)
merged_fresh = []
for start, end in fresh:
    if not merged_fresh or merged_fresh[-1][1] < start:
        merged_fresh.append([start, end])
    else:
        merged_fresh[-1][1] = max(merged_fresh[-1][1], end)
sum(end - start + 1 for start, end in merged_fresh)
```
