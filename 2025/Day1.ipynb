```python
from aocd import get_data
import numpy as np
raw_data = get_data(day=1, year=2025)
```

## Part 1

```python
data = np.array([int(i.replace("L", "-").replace("R", "+")) for i in raw_data.splitlines()])
data = np.insert(data, 0, 50)
data_cumsum = np.cumsum(data)
x1, x2 = np.divmod(data_cumsum, 100)
print(int((x2 == 0).sum()))
```

## Part 2

```python
a = data_cumsum[:-1]  # start of each move
b = data_cumsum[1:]   # end of each move

# Range of clicked values (excluding start, including end)
lo = np.minimum(a + 1, b)
hi = np.maximum(a - 1, b)

clicks = hi // 100 - (lo - 1) // 100

print(int(clicks.sum()))
```
