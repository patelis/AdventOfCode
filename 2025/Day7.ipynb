```python
from aocd import get_data
from collections import Counter
raw_data = get_data(day=7, year=2025)
data = [list(line) for line in raw_data.splitlines()]
```

## Part 1

```python
counter = 0
next_beam_positions = set()
next_beam_positions.add(data[0].index("S"))

for row in data[1:]:
    current_beam_positions = next_beam_positions
    next_beam_positions = set()
    for beam in current_beam_positions:
        if row[beam] == "^" and beam - 1 >= 0 and beam+1 < len(row):
            next_beam_positions.add(beam-1)
            next_beam_positions.add(beam+1)
            counter += 1
        else:
            next_beam_positions.add(beam)
print(counter)
```

## Part 2

```python
next_timelines = Counter([data[0].index("S")])

for row in data[1:]:
    current_timelines = next_timelines
    next_timelines = Counter()
    for column, count in current_timelines.items():
        if row[column] == "^" and column - 1 >= 0 and column+1 < len(row):
            next_timelines[column-1] += count
            next_timelines[column+1] += count
        else:
            next_timelines[column] += count
print(next_timelines.total())
```
