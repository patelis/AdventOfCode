```python
from aocd import get_data
import numpy as np
raw_data = get_data(day=2, year=2025)
```

## Part 1

```python
data = [i.split("-") for i in raw_data.split(",")]
data = [list(map(int, i)) for i in data]

list_integers = []
list_strings = []
for li in data:
    for i in range(li[0], li[1]+1):
        list_integers.append(i)
        list_strings.append(str(i))
        
index = [st[:len(st)//2] == st[len(st)//2:] for st in list_strings]
part1 = int(np.array(list_integers)[index].sum())
part1
```

## Part 2

```python
index = [st in (st + st)[1:-1] for st in list_strings]
part2 = int(np.array(list_integers)[index].sum())
part2
```
