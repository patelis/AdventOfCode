```python
from aocd import get_data
import numpy as np
from scipy import signal
raw_data = get_data(day=4, year=2025)
data = np.array([list(line) for line in raw_data.replace(".", "0").replace("@", "1").splitlines()]).astype(int)
kernel = np.ones((3, 3), dtype=int)
```

## Part 1

```python
summed_neighbors = signal.convolve2d(data, kernel, mode='same', boundary='fill', fillvalue=0)
actual_sum = summed_neighbors - data
part1 = int(np.sum((actual_sum < 4) & (data == 1)))
print(part1)
```

## Part 2

```python
part2 = 0
while True:
    summed_neighbors = signal.convolve2d(data, kernel, mode='same', boundary='fill', fillvalue=0)
    actual_sum = summed_neighbors - data
    part2 += int(np.sum((actual_sum < 4) & (data == 1)))
    new_data = data.copy()
    new_data[(actual_sum < 4) & (data == 1)] = 0
    if np.array_equal(new_data, data):
        break
    data = new_data
print(part2)
```
